```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import config
from omdbapi.movie_search import GetMovie
import warnings
warnings.filterwarnings('ignore')
```

```python
movies_path = f'{config.processed_data_path}/movies.csv'
```

```python
API_KEY = 'c25ecc9c'
```

```python
movies_data = pd.read_csv(movies_path)
movies_data.head()
```

```python
movie_titles = [' '.join(title.split(' ')[:-1]) for title in movies_data['title']]
```

```python
movies_search_engine = GetMovie(api_key=API_KEY)
```

```python
attributes = []
for movie in movie_titles: 
    m = movies_search_engine.get_movie(title=movie)
    attributes.append(m)
```

```python
plots = []
for movie_attr in attributes:
    if type(movie_attr) is not dict:
        plots.append(movie_attr)
    else:
        plots.append(movie_attr['plot'])
```

```python
len(plots)
```

```python
movie_plots = pd.DataFrame(plots, columns={'plot'})
```

```python
movie_plots['movie_id'] = np.arange(len(plots))
```

```python
movie_plots.to_csv(f'{config.processed_data_path}/movie_plots.csv', index=False)
```
